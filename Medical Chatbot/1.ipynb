```python
import requests

def query_drugbank_api(drug_name):
    url = f"https://api.drugbank.com/v1/drugs?name={drug_name}"
    headers = {"Authorization": "Bearer YOUR_API_KEY"}
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()
    return None

# Test the function
print(query_drugbank_api("Metformin"))
```

```python
import requests

# Define the RxNorm API base URL
RX_NORM_API_URL = "https://rxnav.nlm.nih.gov/REST"

# Function to get RxCUI by drug name
def get_rxcui(drug_name):
    url = f"{RX_NORM_API_URL}/rxcui.json?name={drug_name}&search=2"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        if "idGroup" in data and "rxnormId" in data["idGroup"]:
            return data["idGroup"]["rxnormId"][0]  # Return the first RxCUI
    return None

# Function to get drug details by RxCUI
def get_drug_details(rxcui):
    url = f"{RX_NORM_API_URL}/rxcui/{rxcui}/allinfo.json"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    return None

# Test the functions
drug_name = "Metformin 500 mg oral tablet"
rxcui = get_rxcui(drug_name)
if rxcui:
    print(f"RxCUI for {drug_name}: {rxcui}")
    details = get_drug_details(rxcui)
    if details:
        print("Drug Details:")
        print(details)
else:
    print(f"No RxCUI found for {drug_name}.") 
```

```python
def get_related_concepts(rxcui):
    url = f"{RX_NORM_API_URL}/rxcui/{rxcui}/allrelated.json"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        if "allRelatedGroup" in data and "conceptGroup" in data["allRelatedGroup"]:
            return data["allRelatedGroup"]["conceptGroup"]
    return None

# Test the function
related_concepts = get_related_concepts(rxcui)
if related_concepts:
    print("Related Concepts:")
    for group in related_concepts:
        if "conceptProperties" in group:
            for concept in group["conceptProperties"]:
                print(f"{concept['tty']}: {concept['name']}")
else:
    print("No related concepts found.")
```

```python
def get_ndcs(rxcui):
    url = f"{RX_NORM_API_URL}/rxcui/{rxcui}/ndcs.json"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        if "ndcGroup" in data and "ndcList" in data["ndcGroup"]:
            return data["ndcGroup"]["ndcList"]["ndc"]
    return None

# Test the function
ndcs = get_ndcs(rxcui)
if ndcs:
    print("NDCs:")
    for ndc in ndcs:
        print(ndc)
else:
    print("No NDCs found.")
```

```python
from flask import Flask, request, jsonify
import requests

app = Flask(__name__)

RX_NORM_API_URL = "https://rxnav.nlm.nih.gov/REST"

@app.route("/drug-info", methods=["GET"])
def get_drug_info():
    drug_name = request.args.get("drug")
    if not drug_name:
        return jsonify({"error": "Please provide a drug name."}), 400
    
    rxcui = get_rxcui(drug_name)
    if rxcui:
        details = get_drug_details(rxcui)
        if details:
            return jsonify({
                "name": details.get("name"),
                "synonym": details.get("synonym"),
                "termType": details.get("tty")
            })
    return jsonify({"error": "No details found for this drug."}), 404

if __name__ == "__main__":
    app.run(debug=True)
```

```python
# Example response from the RxNorm API
related_drugs = [
    {"tty": "BN", "name": "Glucophage"},
    {"tty": "DF", "name": "Oral Tablet"},
    {"tty": "DFG", "name": "Oral Product"},
    {"tty": "DFG", "name": "Pill"},
    {"tty": "IN", "name": "metformin"},
    {"tty": "PIN", "name": "metformin hydrochloride"},
    {"tty": "SBD", "name": "metformin hydrochloride 500 MG Oral Tablet [Glucophage]"},
    {"tty": "SBDC", "name": "metformin hydrochloride 500 MG [Glucophage]"},
    {"tty": "SBDF", "name": "metformin Oral Tablet [Glucophage]"},
    {"tty": "SBDFP", "name": "metformin hydrochloride Oral Tablet [Glucophage]"},
    {"tty": "SBDG", "name": "Glucophage Oral Product"},
    {"tty": "SBDG", "name": "Glucophage Pill"},
    {"tty": "SCD", "name": "metformin hydrochloride 500 MG Oral Tablet"},
    {"tty": "SCDC", "name": "metformin hydrochloride 500 MG"},
    {"tty": "SCDF", "name": "metformin Oral Tablet"},
    {"tty": "SCDFP", "name": "metformin hydrochloride Oral Tablet"},
    {"tty": "SCDG", "name": "metformin Oral Product"},
    {"tty": "SCDG", "name": "metformin Pill"},
    {"tty": "SCDGP", "name": "metformin hydrochloride Pill"},
    {"tty": "SCDGP", "name": "metformin hydrochloride Oral Product"}
]

# Organize related drugs by term type
related_drugs_by_type = {}
for drug in related_drugs:
    tty = drug["tty"]
    name = drug["name"]
    if tty not in related_drugs_by_type:
        related_drugs_by_type[tty] = []
    related_drugs_by_type[tty].append(name)

# Generate a layman-friendly response
def generate_layman_response(related_drugs_by_type):
    response = []
    
    # Brand Names
    if "BN" in related_drugs_by_type:
        brand_names = related_drugs_by_type["BN"]
        response.append(f"This drug is also sold under the brand name(s): **{', '.join(brand_names)}**.")
    
    # Ingredients
    if "IN" in related_drugs_by_type:
        ingredients = related_drugs_by_type["IN"]
        response.append(f"The active ingredient in this drug is **{', '.join(ingredients)}**.")
    
    # Precise Ingredients
    if "PIN" in related_drugs_by_type:
        precise_ingredients = related_drugs_by_type["PIN"]
        response.append(f"The precise form of the active ingredient is **{', '.join(precise_ingredients)}**.")
    
    # Dose Forms
    if "DF" in related_drugs_by_type:
        dose_forms = related_drugs_by_type["DF"]
        response.append(f"This drug is available as **{', '.join(dose_forms)}**.")
    
    # Dose Form Groups
    if "DFG" in related_drugs_by_type:
        dose_form_groups = related_drugs_by_type["DFG"]
        response.append(f"It comes in forms like **{', '.join(dose_form_groups)}**.")
    
    return "\n".join(response)

# Generate and display the response
layman_response = generate_layman_response(related_drugs_by_type)
print(layman_response)
```
